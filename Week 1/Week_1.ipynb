# 1. WAP to convert temperature from Celsius to Fahrenheit.





```python
def cel_to_fah(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit
def fah_to_cel(fahrenheit):
    celsius = (fahrenheit - 32) * 5/9
    return celsius


celcius = float(input("Enter temperature in Celsius: "))
fahrenheit = cel_to_fah(celcius)
print("Temperature in Fahrenheit:", fahrenheit)

fahrenheit = float(input("Enter temperature in Fahrenheit: "))
celsius = fah_to_cel(fahrenheit)
print("Temperature in Celsius:", celsius)
```

# 2. WAP to swap two numbers

```python
def swap_numbers(a, b):
  a, b = b, a
  return a, b

a = 10
b = 20
a, b = swap_numbers(a, b)
print("a =", a)
print("b =", b)
```

# 3. WAP to check whether a given year is leap year or not

```python
def leap_year(year):
  if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
    return True
  else:
    return False

year = int(input("Enter a year: "))
if leap_year(year):
  print(year, "is a leap year.")
else:
  print(year, "is not a leap year.")
```

# 4. WAP to display reverse of a number

```python
def reverse_number(num):
  reversed_num = 0
  while num > 0:
    digit = num % 10
    reversed_num = reversed_num * 10 + digit
    num //= 10
  return reversed_num

num = int(input("Enter a number: "))
print("Reversed number:", reverse_number(num))
```

# 5. WAP to find factors of a given number


```python
def factors(num):
  print("Factors of", num, "are:")
  for i in range(1, num + 1):
    if num % i == 0:
      print(i)

num = int(input("Enter a number: "))
factors(num)
```

# 6. WAP to generate prime number series up to n


```python
def prime_series(start, end):
  print("Prime numbers between", start, "and", end, "are:")
  for num in range(start, end + 1):
    if num > 1:
      for i in range(2, num):
        if (num % i) == 0:
          break
      else:
        print(num)

start = int(input("Enter the starting number: "))
end = int(input("Enter the ending number: "))
prime_series(start, end)
```

# 7. Write a program which makes use of function to display all such numbers which are divisible by 7 but are not a multiple of 5, between 1000 and 2000.


```python
def divisible_by_7_not_multiple_of_5(start, end):
  print("Numbers divisible by 7 but not a multiple of 5 between", start, "and", end, "are:")
  for num in range(start, end + 1):
    if num % 7 == 0 and num % 5 != 0:
      print(num, end=", ")

start = 1000
end = 2000
divisible_by_7_not_multiple_of_5(start, end)
```

# 8. WAP to check whether a number is a palindrome or not


```python
def palindrome(num):
  temp = num
  reversed_num = 0
  while temp > 0:
    digit = temp % 10
    reversed_num = reversed_num * 10 + digit
    temp //= 10
  if num == reversed_num:
    return True
  else:
    return False

num = int(input("Enter a number: "))
if palindrome(num):
  print(num, "is a palindrome.")
else:
  print(num, "is not a palindrome.")
```

# 9. WAP to check whether
*   is a perfect number
*   is an Armstrong number



```python
def perfect_number(num):
  sum = 0
  for i in range(1, num):
    if num % i == 0:
      sum += i
  if sum == num:
    return True
  else:
    return False

def armstrong_number(num):
  sum = 0
  temp = num
  while temp > 0:
    digit = temp % 10
    sum += digit ** 3
    temp //= 10
  if sum == num:
    return True
  else:
    return False

num = int(input("Enter a number: "))
if perfect_number(num):
  print(num, "is a perfect number.")
else:
  print(num, "is not a perfect number.")

if armstrong_number(num):
  print(num, "is an Armstrong number.")
else:
  print(num, "is not an Armstrong number.")
```

# 10. WAP to generate the Fibonacci series upto n.

```python
def fibonacci(n):
  a = 0
  b = 1
  if n <= 0:
    print("Invalid input.")
  elif n == 1:
    print(a)
  else:
    print(a, end=" ")
    print(b, end=" ")
    for i in range(2, n):
      c = a + b
      print(c, end=" ")
      a = b
      b = c

n = int(input("Enter the number of terms: "))
fibonacci(n)
```
