# 3. Write a Java program to change temperature from Celsius to Fahrenheit.


```python
def cel_to_fah(celsius):
    return (celsius * 9/5) + 32
def fah_to_cel(fahrenheit):
    return (fahrenheit - 32) * 5/9

celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = float(input("Enter temperature in Fahrenheit: "))

print("Celsius to Fahrenheit:", cel_to_fah(celsius))
print("Fahrenheit to Celsius:", fah_to_cel(fahrenheit))
```

# 4. Write a Java program to find area and perimeter of a rectangle.

```python
def area_rectangle(length, width):
    return length * width
def perimeter_rectangle(length, width):
    return 2 * (length + width)

length = float(input("Enter length of rectangle: "))
width = float(input("Enter width of rectangle: "))

print("Area of rectangle:", area_rectangle(length, width))
print("Perimeter of rectangle:", perimeter_rectangle(length, width))
```

# 5. Write a Java program to find area and perimeter of a circle.

```python
def area_circle(radius):
    return 3.14 * radius * radius
def perimeter_circle(radius):
    return 2 * 3.14 * radius

radius = float(input("Enter radius of circle: "))

print("Area of circle:", area_circle(radius))
print("Perimeter of circle:", perimeter_circle(radius))
```

# 8. Write a Java program to find maximum of three numbers.

```python
def find_maximum(num1, num2, num3):
    # return max(num1, num2, num3)
    if num1 >= num2 and num1 >= num3:
        return num1
    elif num2 >= num1 and num2 >= num3:
        return num2
    else:
        return num3

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))

print("Maximum of three numbers:", find_maximum(num1, num2, num3))
```

# 12. Write a Java program to check whether a year is leap year or not.


```python
def check_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
        return True
    else:
        return False

year = int(input("Enter a year: "))

if check_leap_year(year):
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

# 13. Write a Java program for following grading system.
Note: Percentage>=90% : Grade A Percentage>=80% : Grade B Percentage>=70% : Grade C Percentage>=60% : Grade D Percentage>=40% : Grade E Percentage<40% : Grade F


```python
def calculate_grade(percentage):
    if percentage >=90:
        return "A"
    elif percentage >=80:
        return "B"
    elif percentage >=70:
        return "C"
    elif percentage >=60:
        return "D"
    elif percentage >=40:
        return "E"
    else:
        return "F"

total_marks = float(input("Enter total marks: "))
obtained_marks = float(input("Enter obtained marks: "))

percentage = (obtained_marks / total_marks) * 100

grade = calculate_grade(percentage)

print("Percentage:", percentage, "\nGrade:", grade)
```

# 14. Write a Java program to check whether a number is divisible by 5 or not.

```python
def divisible_by_5(number):
    if number % 5 == 0:
        return True
    else:
        return False

number = float(input("Enter a number: "))
result = divisible_by_5(number)

if result:
    print(number, "is divisible by 5.")
else:
    print(number, "is not divisible by 5.")
```
