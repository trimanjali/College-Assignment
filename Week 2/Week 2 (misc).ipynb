# 1. Write a Java program to check whether a number is Buzz or not.

```python
# 1. Write a Java program to check whether a number is Buzz or not.
def check_buzz(num):
    if num % 7 == 0:
        return True
    else:
        return False

num = int(input("Enter a number: "))
if check_buzz(num):
    print(f"{num} is a Buzz number.")
else:
    print(f"{num} is not a Buzz number.")
```

# 2. Write a Java program to calculate factorial of 12.

```python
# 2. Write a Java program to calculate factorial of 12.
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

num = 12
print(factorial(num))
```

# 3. Write a Java program for Fibonacci series.

```python
# 3. Write a Java program for Fibonacci series.
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

num = 3
for i in range(num):
    print(fibonacci(i), end=",")
```

# 4. Write a Java program to reverse a number.

```python
# 4. Write a Java program to reverse a number.
def reverse_number(num):
    reversed_num = 0
    while num > 0:
        digit = num % 10
        reversed_num = reversed_num * 10 + digit
        num //= 10
    return reversed_num

num = 12345
print(reverse_number(num))
```

# 5. Write a Java program to calculate the sum of natural numbers up to a certain range.

```python
# 5. Write a Java program to calculate the sum of natural numbers up to a certain range.
def sum(n):
    sum = 0
    for i in range(1, n + 1):
        sum += i
    return sum

num = 10
print(sum(num))
```

# 6. Write a Java program to print all multiple of 10 between a given interval.

```python
# 6. Write a Java program to print all multiple of 10 between a given interval.
def multiple(start, end):
    for i in range(start, end + 1):
        if i % 10 == 0:
            print(i, end=",")

start = 10
end = 100
multiple(start, end)
```

# 10. Write a Java program to find HCF of two Numbers.

```python
# 10. Write a Java program to find HCF of two Numbers.
def hcf(a, b):
    if b == 0:
        return a
    else:
        return hcf(b, a % b)

num1 = 12
num2 = 18
print(hcf(num1, num2))
```

# 11. Write a Java program to find LCM of two Numbers.

```python
# 11. Write a Java program to find LCM of two Numbers.
def lcm(a, b):
    return (a * b) // hcf(a, b)

num1 = 12
num2 = 18
print(lcm(num1, num2))
```

# 12. Write a Java program to count the number of digits of an integer.

```python
# 12. Write a Java program to count the number of digits of an integer.
def count_digits(num):
    count = 0
    while num > 0:
        num //= 10
        count += 1
    return count

num = 12345
print(count_digits(num))
```

# 13. Write a Java program to calculate the exponential of a number.

```python
# 13. Write a Java program to calculate the exponential of a number.
def exponential(base, exponent):
    result = 1
    for i in range(exponent):
        result *= base
    return result

base = 2
exponent = 3
print(exponential(base, exponent))
```
